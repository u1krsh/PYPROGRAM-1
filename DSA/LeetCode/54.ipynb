```python
matrix = [[1,2,3],[4,5,6],[7,8,9]]
res = []
class Solution(object):
    def spiralOrder(self, matrix):
        while True:
            res = []
    
            def fin_check():
                if len(matrix) == 0:
                    return True
            while True:
        # step1 append elements of first non-empty list
                res.extend(matrix[0])
                matrix.pop(0)
                matrix = [item for item in matrix if len(item)>0]
                if fin_check() : return(res)
        
        # step 2 append last element of all lists in order
                for col in matrix : 
                    res.append(col[-1])
                    col.pop(-1)
                matrix = [item for item in matrix if len(item)>0]
        
                if fin_check():return(res)
        
        #step 3 append reverse of last list
                res.extend(matrix[-1][::-1])
                matrix.pop(-1)
                matrix = [item for item in matrix if len(item)>0]
        
                if fin_check():return(res)
        
        #step 4 append first element of all lists in reverse order
                for col in reversed(matrix) : 
                    res.append(col[0])
                    col.pop(0)
                    matrix = [item for item in matrix if len(item)>0]
            if fin_check():return(res)

print(spiral(matrix))
```
